#### Função para a tabela de frequência

```python
def tabela_frequencia(variavel):
    res = pd.DataFrame(dados[variavel].value_counts()) 
    res.columns = ["Quantidade"]
    
    res["Percentual"] = res.Quantidade / res.Quantidade.sum() * 100
    res["Percentual"] = res["Percentual"].apply(lambda x: "{0:.2f}%".format(x)) #formatar x para todo x em Percentual

    return res
```

#### Quantidade de empréstimos por tipo de vínculo

```python
dados["tipo_vinculo_usuario"].unique()
```

```python
dados_vinculo = tabela_frequencia("tipo_vinculo_usuario")

dados_vinculo
```
