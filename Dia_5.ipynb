### Importando dados

```python
dados = Dia_4.dados

dados.head()
```

### Função para plotar boxplot

```python
def plot_boxplot(df, x, y, titulo, subtitulo, palette=sns.color_palette("hls", 8)):
    plt.figure(figsize=(16,10))
    ax = sns.boxplot(x=x, y=y, data=df, palette=palette)
    
    plt.ylim(0,max(df[y])*1.1) # limite superior do eixo y
    ax.set(xlabel=None,ylabel=None) # labels dos eixos
    
    ax.set_title(titulo,size=20,loc='left',weight='bold') # título
    ax.text(s=subtitulo,x=-0.5,y=max(df[y])*1.118,fontsize=18, ha='left',color='gray') # subtítulo
    
    plt.show
```

### Primeira parte: Alunos de graduação

#### Fazendo queries

```python
grad = dados.query("tipo_vinculo_usuario == \"ALUNO DE GRADUAÇÃO\"")
grad["colecao"].value_counts()
```

```python
grad_freq = grad.query("colecao == \"Acervo Circulante\"")

grad_freq
```

```python
grad_freq_mensais = pd.DataFrame()

grad_freq_mensais["ano"] = grad_freq["data_emprestimo"].dt.year
grad_freq_mensais["mês"] = grad_freq["data_emprestimo"].dt.month

grad_freq_mensais = grad_freq_mensais.value_counts().reset_index(name="quantidade")
grad_freq_mensais = grad_freq_mensais.sort_values(by=["ano", "mês"]).reset_index(drop=True)

grad_freq_mensais
```

#### Plotando boxplot

```python
plot_boxplot(grad_freq_mensais,
             "ano",
             "quantidade",
             "Distribuição dos empréstimos mensais\n",
             "Alunos de graduação",)
```
