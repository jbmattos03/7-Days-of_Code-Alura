## Dia 1: Importando dados

```python
import pandas as pd
import glob
```

### Parte 1: Empréstimos

#### Pegando tabelas de empréstimos

```python
caminho = r'C:\Users\yello\7_Days_of_Code_Alura-Python-Pandas\Dia_1-Importando_dados\Datasets\dados_emprestimos'
tabelas = glob.glob(caminho + '/*.csv')

lista_emprestimos = []

for tabela in tabelas:
    try:
        lista_emprestimos.append(pd.read_csv(tabela))
    except Exception as e:
        print(f'Erro ao ler tabela {tabela}: {e}')

#lista_emprestimos
```

```python
print(len(lista_emprestimos))
```

#### Visualizando formato das tabelas

```python
lista_emprestimos[0].head()
```

```python
lista_emprestimos[0].shape
```

#### Concatenando tabelas em um só dataframe

```python
dados_emprestimos = pd.concat(lista_emprestimos, ignore_index=True)

dados_emprestimos.shape
```

```python
dados_emprestimos.head()
```

```python
dados_emprestimos.tail()
```

```python
dados_emprestimos.value_counts()
```

#### Limpando dados

```python
dados_emprestimos = dados_emprestimos.drop_duplicates(ignore_index=True)

dados_emprestimos.shape
```
