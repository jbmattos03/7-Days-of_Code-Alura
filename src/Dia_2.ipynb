#### Adicionando classe CDU

```python
lista_classes = []
```

```python
for loc in dados["localizacao"]:
    if 0 <= loc <= 99:
        lista_classes.append("Generalidades, ciência e conhecimento")
    elif 100 <= loc <= 199:
        lista_classes.append("Filosofia e psicologia")
    elif 200 <= loc <= 299:
        lista_classes.append("Religião")
    elif 300 <= loc <= 399:
        lista_classes.append("Ciências sociais")
    elif 400 <= loc <= 499:
        lista_classes.append("Classe vaga")
    elif 500 <= loc <= 599:
        lista_classes.append("Matemática e ciências naturais")
    elif 600 <= loc <= 699:
        lista_classes.append("Ciências aplicadas")
    elif 700 <= loc <= 799:
        lista_classes.append("Belas artes")
    elif 800 <= loc <= 899:
        lista_classes.append("Linguagem, língua, linguística")
    elif 900 <= loc <= 999:
        lista_classes.append("Geografia, biografia, história")
```

```python
dados["classe_CDU"] = lista_classes

dados.head()
```
