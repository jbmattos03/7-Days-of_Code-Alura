### Concatenando tabelas

#### Tratando dados: Excel

```python
ate_2010.columns = ["matricula_ou_siape", "tipo_vinculo_usuario", "curso"]
ate_2010.drop(index=0, axis=0, inplace=True)

ate_2010
```

```python
apos_2010.columns = ["matricula_ou_siape", "tipo_vinculo_usuario", "curso"]
apos_2010.drop(index=0, axis=0, inplace=True)

apos_2010
```

```python
matriculas = pd.concat([ate_2010, apos_2010], ignore_index=True)
matriculas["matricula_ou_siape"] = matriculas["matricula_ou_siape"].astype("float64").astype("int64")

matriculas
```

```python
dados_usuarios = pd.concat([matriculas, cadastros], ignore_index=True)

dados_usuarios
```

### Criando tabela

```python
emprestimos_mais_2015 = Dia_5.dados.query("tipo_vinculo_usuario == \"ALUNO DE GRADUAÇÃO\" & data_emprestimo > 2015").loc[:,['matricula_ou_siape','data_emprestimo']].reset_index(drop=True)
emprestimos_mais_2015["matricula_ou_siape"] = pd.to_numeric(emprestimos_mais_2015["matricula_ou_siape"], errors='coerce')
emprestimos_mais_2015 = emprestimos_mais_2015.dropna(subset=["matricula_ou_siape"])
emprestimos_mais_2015["matricula_ou_siape"] = emprestimos_mais_2015["matricula_ou_siape"].astype("int64")

emprestimos_mais_2015
```

```python
dados_cursos_sel = dados_usuarios.query("curso == ['BIBLIOTECONOMIA','CIÊNCIAS SOCIAIS','COMUNICAÇÃO SOCIAL','DIREITO','FILOSOFIA','PEDAGOGIA']")
dados_cursos_sel
```

```python
dados_completos = emprestimos_mais_2015.merge(dados_cursos_sel)
dados_completos["data_emprestimo"] = dados_completos["data_emprestimo"].dt.year

dados_completos
```

```python
qtd_emp = pd.DataFrame(dados_completos.iloc[:,[1,3]].value_counts().reset_index())
qtd_emp.columns = ["ano", "curso", "quantidade_emprestimos"]

qtd_emp.head()
```

```python
qtd_emp_pivot = qtd_emp.pivot_table(
        index = 'curso',
        columns = 'ano',
        values = 'quantidade_emprestimos',
        fill_value = '-',
        aggfunc= 'sum',
        margins = True,
        margins_name = 'total',
)
qtd_emp_pivot
```
