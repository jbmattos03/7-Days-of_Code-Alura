### Filtrando alunos de pós-graduação

```python
dados_pos = dados_usuarios.query("tipo_vinculo_usuario == \"ALUNO DE PÓS-GRADUAÇÃO\"").reset_index(drop=True)

dados_pos.head()
```

### Filtrando dados a partir de 2017

```python
dados_2017_mais = dados.loc[:, ["matricula_ou_siape", "data_emprestimo"]].query("data_emprestimo > 2017").reset_index(drop=True)
dados_2017_mais["data_emprestimo"] = dados_2017_mais["data_emprestimo"].dt.year

dados_2017_mais.head()
```

```python
dados_pos['matricula_ou_siape'] = dados_pos['matricula_ou_siape'].astype(str)
dados_2017_mais_curso = dados_2017_mais.merge(dados_pos, on='matricula_ou_siape')

dados_2017_mais_curso.head()
```

```python
quantidade_ano_curso = dados_2017_mais_curso.iloc[:, [1, 3]].value_counts().reset_index()
quantidade_ano_curso.columns = ["ano", "curso", "emprestimos"]

quantidade_ano_curso.head()
```

```python
quantidade_ano_curso_pivot = quantidade_ano_curso.pivot(
        index = "curso",
        columns = "ano",
        values = "emprestimos",
)
quantidade_ano_curso_pivot
```

### Comparando emprestimos por ano

#### Criando função para calcular diferença percentual

```python
def diferenca_percentual_ano_anterior(x, y):
  return round(((x / y * 100) - 100), 2)
```

#### 2017-2018

```python
diferenca_2018 = pd.DataFrame(diferenca_percentual_ano_anterior(quantidade_ano_curso_pivot[2018], quantidade_ano_curso_pivot[2017]).apply(lambda x: f"{x}%"))
diferenca_2018.columns = ["diferença 2017-2018"]

diferenca_2018
```

#### 2018-2019

```python
diferenca_2019 = pd.DataFrame(diferenca_percentual_ano_anterior(quantidade_ano_curso_pivot[2019], quantidade_ano_curso_pivot[2018]).apply(lambda x: f"{x}%"))
diferenca_2019.columns = ["diferença 2018-2019"]

diferenca_2019
```

#### 2019-2022

```python
diferenca_2022 = pd.DataFrame(diferenca_percentual_ano_anterior(quantidade_ano_curso_pivot[2022], quantidade_ano_curso_pivot[2019]).apply(lambda x: f"{x}%"))
diferenca_2022.columns = ["diferença 2019-2022"]

diferenca_2022
```

```python
dados_finais = pd.DataFrame(
    {
        "2017-2018": diferenca_2018["diferença 2017-2018"],
        "2018-2019": diferenca_2019["diferença 2018-2019"],
        "2019-2022": diferenca_2022["diferença 2019-2022"]
    }
)

dados_finais.index = dados_finais.index.str.capitalize()

dados_finais
```

### Traduzindo tabela para HTML

```python
def highlight(val):
    color = 'green' if int(''.join(re.findall(r'\d+', val))) > 0 else 'blue' if int(''.join(re.findall(r'\d+', val))) < 0 else 'black'
    return f'color: {color}'
```

```python
dados_finais_coloridos = dados_finais.style.applymap(highlight, subset=["2017-2018", "2018-2019", "2019-2022"])
# FutureWarning: Styler.applymap has been deprecated. Use Styler.map instead
```

```python
html = dados_finais_coloridos.to_html()
```

```python
with open('dados_finais.html', 'w') as f:
    f.write(html)
```
